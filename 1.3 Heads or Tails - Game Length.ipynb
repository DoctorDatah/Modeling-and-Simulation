# Heads or Tails - Winning Game Length

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
nsim = 10000
ngames= 0
choices = (1,-1)

target = 5 # Winning Threshold
current = 0 # Stating Number
winning_length = []
```

```python
while(current<target and ngames<nsim):
    ngames += 1
    current += np.random.choice(choices)
    winning_length.append(current)
   
```

```python
plt.plot(winning_length)
plt.xlabel('Number of Runs')
plt.ylabel('Winnings')

if(ngames >= nsim):
    print("Did Not reach the target in " + str(nsim) +" runs")
else:
    print("Target achived in " + str(len(winning_length)) +" runs")
```
