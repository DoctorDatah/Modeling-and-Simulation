```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Assume we have unfair die with Head probability = 0.51 and Tail with 0.49

```python
heads_proba = 0.51
```

```python
np.random.rand(10000, 10) # 10000 trials in each od 10 simulations # uniformaly distributed
```

```python
coin_tosses = (np.random.rand(10000, 10) < heads_proba).astype(np.int32)
```

```python
coin_tosses.shape
```

```python
np.cumsum(coin_tosses, axis=0).shape
```

```python
np.arange(1, 10001).reshape(-1, 1).shape
```

a1 dim = [10000, 10]  
a2 dim = [10000, 1] 

a1 / a2 means each row (10 values in a row ) in divided by corresponding row in  a2 with only contains 1 value 

```python
cumulative_heads_ratio = np.cumsum(coin_tosses, axis=0) / np.arange(1, 10001).reshape(-1, 1)
```

```python
cumulative_heads_ratio.shape
```

```python
plt.figure(figsize=(8,3.5))
plt.plot(cumulative_heads_ratio)
plt.plot([0, 10000], [0.51, 0.51], "k--", linewidth=2, label="51%")
plt.plot([0, 10000], [0.5, 0.5], "k-", label="50%")
plt.xlabel("Number of coin tosses")
plt.ylabel("Heads ratio")
plt.legend(loc="lower right")
plt.axis([0, 10000, 0.42, 0.58])
plt.show()
```

These 10 columns are 10 independent simulations
